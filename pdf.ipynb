## 2. Parse PDF File

In this section, we will perform the following tasks:

* Read the PDF file and extract the necessary tables in the PDF file
* Extract and examine data


#### 2.1 Read PDF file and Extract tables

```python
# health.pdf Is instantiated as pdfFile
pdfFile = 'health.pdf'

# File is read
pdfobj = open(pdfFile, 'rb')

# get_tables from the pdftables library is used to parse the pdfobj and find all tables
tables = get_tables(pdfobj)
```

#### 2.2 Extract and examine tables

We can see that the lay out of the parsed rows is such that the first few rows on each page have the titles and column headings spread out in an unreadable fashion. With the first row that holds the required information of a country (Afghanistan) and the following rows up to the penultimate page; the rows appear to be ordered in roughly the same way, however with closer examination we see that some rows are shorter than others as values have been concatenated. With the last page containing tables we see once again that headers and column names spread out, however the required rows that hold country information beginning with Uganda are even shorter than previous rows, as some row values have been concatenated with the country names and the row values following after also do not imitate the orginal pdf file. There are other small less common errors such as the concatenation of the en dash which was used to represent missing information.

```python
# Uisng a for loop print out the first 20 rows from each page of the PDF object
# ('==========================\n') is used to indicate a new page

for table in tables:
    for row in table[:20]:
        print (row)
    print ('==========================\n')
```

## 3. Clean Data

In this section, we will perform the following tasks:
* Separate the Country name from data values
* Separate all concatenated values including en dashes and decimal digits
* Remove x values
* Deal with any other patterns that are not common but affect the cleanliness of the data




#### 3.1 Separate country name and row values

```python
# First examine the country names that have been cocatenated with row values to be thorough
# what exactly are the 1st and 2nd values of each row?

for table in tables:
    for row in table[5:]:
        print(row[0])
        print(row[1])
```

```python
# iterate through each relevant row in tables using a for loop

for table in tables:
    for row in table[5:]:
        # using regex split function, split the first row value of any row that has country name concatenated with integers
        # split where digits begin
        string = re.split('(\d+)', row[0])
        # if row[0] holds value of country name + integers then string object will be a list of more than one item
        if len(string) > 1:
            # perform the below funcitons if if test is passed
            row[0] = string[0]
            # row[0] is now country name
            row[1] = string[1] + row[1]
            # row [1] is now the split integers concatenated with the previous value of row[1] 
        if row[0] == '':
            break
            # Now that all country names have separated from any decimal digits, we break when we reach row with Zimbabwe for country name or row[0]
        if row[0] == 'Zimbabwe':
            print(row)
            break
        print(row)
        # signifies end of page
    print('==========================\n')
```

#### 3.2 Formulate the code to clean data of concatenated values, en dashes, "x"s , and other unwated patterns

```python
# Uisng for loops and while loops iterate over tables and rows to clean the data
# Each row is printed with alterations to show current state to altered state and change in row length to show that the ultimate goal is to reach 23 values for each row with values in correct places
# The functions within the while loops are ordered in a specific way so that each regex pattern can be matched with an altered row of needed 

i = 1
j = 1

for table in tables:
    # iterate over relevant lines
    for row in table[5:]:
        while i < len(row):
            # Venezuela alone had a particular pattern where one value of the row followed the pattern "38x– 
            if row[0] == 'Venezuela (Bolivarian Republic of)':
                # to ensure that no other values in other rows or other row values in the Venzuela row would be altered the following regex pattern was used and the row was specified 
                if re.findall(r'3?.8?x?.–$', row[i]):
                    # the pattern indicates the start and end of the pattern as well the number of occurrences, find all values that fit this pattern 
                    split = re.findall(r'\d+.x', row[i])
                    # split will equal to "38x"
                    split1 = re.split(r'\d+.x?', row[i])
                    # split1 will equal to ["–", ""]
                    row = row[:i] + split + split1[1:] + row[i+1:]
                    # build a new row that includes the en dash to represent missing information and to make sure that the row ultimately contains 23 values
                    print(row)
                    print(len(row))
            # if the len row[i] is more than 3 digits, then we look for specific patterns, either values that are "100", or "0"s that occur more than one time or values that need to be divided into a length of 2 have been concatenated
            if len(row[i]) > 3:
                # find all patterns with "100", a "0" that occurs one time, and or divide string into decimal digits of length 2
                row1 = re.findall(r'(1[0]{2}|0{1}|[0-9]{,2})', row[i])
                # build a new row with separated strings
                row = row[:i] + row1[:-1] + row[i+1:]
                print(row)
                print(len(row))
            # if row[i] has an x in string then split x and build new row without x value
            if re.findall(r'x', row[i]):
                y = row[i].split("x")
                row = row[:i] + y[:-1] + row[i+1:]
                print(row)
                print(len(row))
            # if any non-alphanumeric character concatenated with en dash split string, build new row
            if re.findall(r'\W.–{,1}', row[i]):
                jay = row[i].split('–')
                row = row[:i] + jay[:-1] + row[i+1:]
                print(row)
                print(len(row))
            i += 1
        # This is to specifically tackle Uzbekistan which occurs as "Uzbekistan–"
        if re.findall(r'\w+.–', row[0]):
            row1 = row[0].split('–')
            # new row is built with country name and dash separated, once again ensuring that each row has 23 values
            row = row1 + row[1:]
            print(row)
            print(len(row))
        # a new while loop is used to account for any further cleansing after the above alterations have occured
        # looking through the data that there are a few errors where row values that are greater than three and have concatenated 1 or two row values
        while j < len(row):
            if len(row[j]) == 3:
                # to tackle Ethiopia where one row value is "718" instead of "7" and "18"
                if re.findall(r'(^7{1})', row[j]):
                    split = re.findall(r'7{1}', row[j])
                    split1 = re.split(r'7{1}', row[j])
                    row = row[:j] + split + split1[1:] + row[j+1:]
                # to tackle Liberia: "286" to "28", "6", Guinea Bisau "358" to "35", "8", Ghana "199" to "19", "9" and Madagascar: "166" to "16", "6"
                elif re.findall(r'[896]{1}$', row[j]):
                    split = re.findall(r'[896]{1}$', row[j])
                    split1 = re.split(r'[896]{1}$', row[j])
                    row = row[:j] + split1[:-1] + split + row[j+1:]
                # to alter Ukraine: 00– to 0, 0, –
                elif re.findall(r'^0{2}', row[j]):
                    split = re.findall(r'0{1}', row[j])
                    split1 = re.split(r'0{2}', row[j])
                    row = row[:j] + split + split1[1:] + row[j+1:]
            # to finally separate any row values with decimal digits concatenated with –
            if re.findall(r'\d+.–', row[j]):
                jay = row[j].split('–')
                row = row[:j] + jay + row[j+1:]
                print(row)
                print(len(row))
            j += 1
        i = 1
        j = 1
        if row[0] == '':
            break
        if row[0] == 'Zimbabwe':
            print(row)
            break
        print(row)
        print(len(row))
    print('==========================\n')
```

## 4. Building a Data Frame

In this section, we will perform the following tasks:
* Put all of the above elements together
* Build a Pandas Data Frame
* Alter the Data Frame to meet the requirements of the task



#### 4.1 Put all elements together and create a Pandas Data frame

```python
# Extract data
pdfFile = 'health.pdf'
pdfobj = open(pdfFile, 'rb')
tables = get_tables(pdfobj)

# Alter concatenated country names
for table in tables:
    for row in table[5:]:
        string = re.split('(\d+)', row[0])
        if len(string) > 1:
            row[0] = string[0]
            row[1] = string[1] + row[1]
        if row[0] == '':
            break
        if row[0] == 'Zimbabwe':
            break

i = 1
j = 1

# create a dictionary called data
data = {}

# add cleansing code
for table in tables:
    for row in table[5:]:
        while i < len(row):
            if row[0] == 'Venezuela (Bolivarian Republic of)':
                if re.findall(r'3?.8?x?.–$', row[i]):
                    split = re.findall(r'\d+.x', row[i])
                    split1 = re.split(r'\d+.x?', row[i])
                    row = row[:i] + split + split1[1:] + row[i+1:]
            if len(row[i]) > 3:
                row1 = re.findall(r'(1[0]{2}|0{1}|[0-9]{,2})', row[i])
                row = row[:i] + row1[:-1] + row[i+1:]
            if re.findall(r'x', row[i]):
                y = row[i].split("x")
                row = row[:i] + y[:-1] + row[i+1:]
            if re.findall(r'\W.–{,1}', row[i]):
                jay = row[i].split('–')
                row = row[:i] + jay[:-1] + row[i+1:]
            i += 1
        if re.findall(r'\w+.–', row[0]):
            row1 = row[0].split('–')
            row = row1 + row[1:]
        while j < len(row):
            if len(row[j]) == 3:
                if re.findall(r'(^7{1})', row[j]):
                    split = re.findall(r'7{1}', row[j])
                    split1 = re.split(r'7{1}', row[j])
                    row = row[:j] + split + split1[1:] + row[j+1:]
                elif re.findall(r'[896]{1}$', row[j]):
                    split = re.findall(r'[896]{1}$', row[j])
                    split1 = re.split(r'[896]{1}$', row[j])
                    row = row[:j] + split1[:-1] + split + row[j+1:]
                elif re.findall(r'^0{2}', row[j]):
                    split = re.findall(r'0{1}', row[j])
                    split1 = re.split(r'0{2}', row[j])
                    row = row[:j] + split + split1[1:] + row[j+1:]
            if re.findall(r'\d+.–', row[j]):
                jay = row[j].split('–')
                row = row[:j] + jay + row[j+1:]
            j += 1
        i = 1
        j = 1
        if row[0] == '':
            break
        if row[0] == 'Zimbabwe':
            data[row[0]] = row[1:]
            break
        # build dictionary
        data[row[0]] = row[1:]
# Build data frame
data = pd.DataFrame(data)

data
```

#### 4.2 Alter the Data Frame

```python

# transpose data and create new object
healthPDF = data.T

# using a for loop  replace all en dashes with now empty strings
for i in range(22):
    healthPDF[i].replace(regex=True,inplace=True,to_replace=r'\W',value=r'')
    
# alter index name to be "Country names
healthPDF.index.names = ['Country names']
```
